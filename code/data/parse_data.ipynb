# Create .csv file for each year using population.csv

```python
import pandas as pd
```

```python
populations = pd.read_csv("population.csv")

populations.head()
```

```python
populations.describe()
```

```python
populations.info()
```

```python
for year in range(2010, 2025):
    print(f"Year: {year}")
    print(f"Num examples: {len(populations[populations['year'] == year])}\n")
```

```python
#create new populations file for each year so we can analyze as needed.
for year, year_df in populations.groupby("year"):
    file = f'population{int(year)}.csv'
    year_df.to_csv(file, index=False)
```
